```python
import scholarly
```

```python
search_query = scholarly.search_author('Sanjay Aneja, MD')
```

```python
author = next(search_query).fill()
print(author)
```

```python
publications = [pub.bib['title'] for pub in author.publications]
```

```python
for i in range(len(publications)):
    print(author.publications[i].fill())
```

```python
import string
sample_pub = author.publications[0].fill()
# Title. Authors. Journal. Year
publication_citations = []

error_pubs = []
for i in range(len(publications)):
    try:
        print(i)
        current_pub = author.publications[i]
        authors = current_pub.bib['author']
        authors = ", ".join(authors.split(' and '))
        journal = current_pub.bib['journal']
        journal = string.capwords(journal)
        year = current_pub.bib['year']
        publication_dict = {}
        publication_dict['citation'] = "{}. {}. {}. {}".format(current_pub.bib['title'], authors, journal, year)
        publication_dict['year'] = year
        publication_citations.append(publication_dict)
    except Exception as e:
        error_dict = {}
        error_dict['error'] = e
        error_dict['image_number'] = i
        error_pubs.append(error_dict)
```

```python
imageDirectory="../2019images/Personal-Friends-Final/"
outputPath="publications.html"

import os

f = open(outputPath, "w+")
publication_count = 0

# Write the opening
f.write("<ul class='collection with-header'>")
f.write("<li class='collection-header'><h4>Publications</h4></li>")

for publication in sorted_publication_citations:
    # Add the citation
    citation = publication["citation"]
    f.write("<li class='collection-item'>{}</li>".format(citation))
    publication_count += 1

f.write("</ul>")
        
f.close()
```
