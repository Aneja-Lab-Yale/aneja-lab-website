```python
import scholarly
```

```python
search_query = scholarly.search_author('Sanjay Aneja, MD')
```

```python
author = next(search_query).fill()
print(author)
```

```python
publications = [pub.bib['title'] for pub in author.publications]
```

```python
for i in range(len(publications)):
    print(author.publications[i].fill())
```

```python
sample_pub = author.publications[0].fill()
# Title. Authors. Journal. Year
publication_citations = []

for i in range(len(publications)):
    print(i)
    current_pub = author.publications[i].fill()
    publication_citations.append("{}. {}. {}. {}".format(sample_pub.bib['title'], sample_pub.bib['author'], sample_pub.bib['journal'], sample_pub.bib['year']))
```
